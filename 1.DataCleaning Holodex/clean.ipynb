```python
import pandas as pd
# Membaca data dari CSV
df = pd.read_csv('holodex_data.csv')

# Menampilkan data awal
print("Data Awal:")
df
```

```python
# Menghapus duplikat
df_cleaned = df.drop_duplicates()

# Merapikan format tanggal 'published_at'
df_cleaned['published_at'] = pd.to_datetime(df_cleaned['published_at']).dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
# Memisahkan data berdasarkan 'lang' dan 'duration'
# Filter untuk 'id' dan 'en'
df_filtered = df_cleaned[(df_cleaned['lang'] == 'id') | (df_cleaned['lang'] == 'en')]
```

```python
# Durasi Tertinggi dan Terendah berdasarkan 'lang' sebelum cleaning
print("\nDurasi Tertinggi dan Terendah Berdasarkan 'lang' Sebelum Cleaning:")
print("Durasi Terendah (id):")
print(df[df['lang'] == 'id'].sort_values(by='duration').iloc[0])
print("Durasi Tertinggi (id):")
print(df[df['lang'] == 'id'].sort_values(by='duration', ascending=False).iloc[0])

print("\nDurasi Terendah (en):")
print(df[df['lang'] == 'en'].sort_values(by='duration').iloc[0])
print("Durasi Tertinggi (en):")
print(df[df['lang'] == 'en'].sort_values(by='duration', ascending=False).iloc[0])
```

```python
# Menentukan author name yang akan dihapus
author_to_remove = 'At Da Momentã€ RORIN ã€‘'  # Ganti dengan nama author yang sebenarnya

# Menghapus baris dengan author name tertentu
df_filtered = df_filtered[df_filtered['author_name'] != author_to_remove].copy()

print(f"\nBaris dengan author name '{author_to_remove}' telah dihapus.")
```

```python
# Menampilkan daftar unique author name
unique_authors = df_filtered['author_name'].unique()
print("\nDaftar Unique Author Name:")
for author in unique_authors:
    print(author)

print(f"\nJumlah unique author: {len(unique_authors)}")
```

```python
# Memisahkan data berdasarkan 'lang' dan 'duration'
df_filtered = df_cleaned[(df_cleaned['lang'] == 'id') | (df_cleaned['lang'] == 'en')]

# Menyaring data durasi antara 10 detik sampai 210 detik
df_duration_filtered = df_filtered[(df_filtered['duration'] >= 10) & (df_filtered['duration'] <= 140 )]

# Memisahkan berdasarkan 'lang' dan menyimpan ke dua file terpisah
df_id = df_duration_filtered[df_duration_filtered['lang'] == 'id']
df_en = df_duration_filtered[df_duration_filtered['lang'] == 'en']

# Menyimpan hasil ke file CSV terpisah
df_id.to_csv('holodex_data_id_filtered.csv', index=False)
df_en.to_csv('holodex_data_en_filtered.csv', index=False)

# Menampilkan informasi file yang disimpan
print("Data untuk 'id' dan 'en' berdasarkan durasi 10-210 detik telah disimpan:")
print("File 'holodex_data_id_filtered.csv' untuk 'id'")
print("File 'holodex_data_en_filtered.csv' untuk 'en'")
```
